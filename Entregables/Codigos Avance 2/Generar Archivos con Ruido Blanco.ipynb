```python
import librosa
import numpy as np
import soundfile as sf

def add_white_noise_to_audio(file_path, noise_level=0.005):
    """
    Agrega ruido blanco a un archivo de audio y guarda el resultado.

    Par치metros:
    - file_path: Ruta al archivo de audio original.
    - noise_level: Nivel del ruido blanco a agregar, como una fracci칩n de la amplitud m치xima de la se침al de audio.

    Retorna:
    - y_noisy: Audio con ruido agregado.
    - sr: Tasa de muestreo del audio.
    """
    # Cargar el archivo de audio
    y, sr = librosa.load(file_path, sr=None)

    # Generar ruido blanco
    noise = np.random.normal(0, noise_level * np.amax(y), y.shape)

    # Agregar el ruido al audio original
    y_noisy = y + noise

    return y_noisy, sr

# Ejemplo de uso
file_path = '/content/drive/MyDrive/DataSet_Voice/frase/Total/Enfermos/2610-phrase.wav'
y_noisy, sr = add_white_noise_to_audio(file_path, noise_level=0.005)

# Opcionalmente, guardar el audio con ruido a un nuevo archivo
sf.write('/content/drive/MyDrive/DataSet_Voice/ruido_blanco/2610-phrase.wav', y_noisy, sr)
```

```python
import os
import glob
import librosa
import numpy as np
import soundfile as sf
def add_white_noise_to_audio(file_path, noise_level=0.005):
    y, sr = librosa.load(file_path, sr=None)
    noise = np.random.normal(0, noise_level * np.amax(y), y.shape)
    y_noisy = y + noise
    return y_noisy, sr
```

```python
def process_audio_files(source_folder, target_folder, noise_level=0.005):
    # Crear el directorio de destino si no existe
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)

    # Listar todos los archivos .wav en el directorio fuente
    for file_path in glob.glob(os.path.join(source_folder, '*.wav')):
        print(f"Procesando {file_path}")

        # Agregar ruido blanco al archivo de audio
        y_noisy, sr = add_white_noise_to_audio(file_path, noise_level=noise_level)

        # Construir la ruta del archivo de salida
        base_name = os.path.basename(file_path)
        output_file_path = os.path.join(target_folder, base_name)

        # Guardar el archivo de audio con ruido
        sf.write(output_file_path, y_noisy, sr)
```

```python
source_folder = '/content/drive/MyDrive/DataSet_Voice/frase/Total/Sanos/'
target_folder = '/content/drive/MyDrive/DataSet_Voice/SanTotalWnoise/'
process_audio_files(source_folder, target_folder, noise_level=0.005)
```
